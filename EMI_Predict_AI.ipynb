### Step 3: Advanced Feature Engineering and Data Splitting

##### 3.1 Derived Financial Ratios (Affordability Features)

```python
# --- ASSUMPTION: df_cleaned from Step 1 is available ---
# If you are running this cell separately, you must load your data and run the cleaning/EDA steps first.
# For demonstration, let's assume 'df_cleaned' exists and contains the necessary columns.
# Replace this line with your actual cleaned DataFrame loading if running standalone.
# df_cleaned = pd.read_csv('emi_prediction_dataset_cleaned.csv')

df_fe = df_cleaned.copy()

# 1. Total Monthly Expenses
expense_cols = ['school_fees', 'college_fees', 'travel_expenses', 'groceries_utilities', 'other_monthly_expenses', 'monthly_rent']
df_fe['total_monthly_expenses'] = df_fe[expense_cols].sum(axis=1)

# 2. Total Current Debt Obligation (Existing EMIs + Requested EMI)
# Requested EMI is not a feature, but existing EMI is a current obligation.
df_fe['total_current_obligation'] = df_fe['current_emi_amount']

# 3. Expense-to-Income Ratio
df_fe['expense_to_income_ratio'] = df_fe['total_monthly_expenses'] / df_fe['monthly_salary']
# Handle potential division by zero or infinites
df_fe['expense_to_income_ratio'].replace([np.inf, -np.inf], 0, inplace=True)
df_fe.loc[df_fe['expense_to_income_ratio'] > 1, 'expense_to_income_ratio'] = 1 # Cap at 1 (or 100%)

# 4. Debt-to-Income Ratio (DTI)
df_fe['debt_to_income_ratio'] = df_fe['total_current_obligation'] / df_fe['monthly_salary']
df_fe['debt_to_income_ratio'].replace([np.inf, -np.inf], 0, inplace=True)
df_fe.loc[df_fe['debt_to_income_ratio'] > 1, 'debt_to_income_ratio'] = 1 # Cap at 1 (or 100%)

# 5. Free Cash Flow (Discretionary Income)
# Simple calculation: Income - (Expenses + Existing Debt)
df_fe['free_cash_flow'] = df_fe['monthly_salary'] - (df_fe['total_monthly_expenses'] + df_fe['total_current_obligation'])
df_fe.loc[df_fe['free_cash_flow'] < 0, 'free_cash_flow'] = 0 # Cannot have negative FCF for affordability

print("‚úÖ Derived Financial Ratios Created.")
print(df_fe[['monthly_salary', 'total_monthly_expenses', 'debt_to_income_ratio', 'free_cash_flow']].head())
```

##### 3.2 Risk Scoring and Interaction Features

```python
# 6. Credit Risk Flag (Simple)
# Flag for customers with low credit score AND existing loan obligations
RISK_THRESHOLD = 650 # Using a standard threshold for demonstration
df_fe['low_credit_risk_flag'] = np.where(
    (df_fe['credit_score'] < RISK_THRESHOLD) & (df_fe['existing_loans'] == 'Yes'),
    1, # High Risk
    0  # Low Risk
)

# 7. Employment Stability Score (Interaction Feature)
# A high salary combined with long employment tenure is a sign of stability.
df_fe['employment_stability_score'] = df_fe['monthly_salary'] * df_fe['years_of_employment']

# 8. Emergency Fund Ratio
df_fe['emergency_fund_ratio'] = df_fe['emergency_fund'] / df_fe['monthly_salary']
df_fe['emergency_fund_ratio'].replace([np.inf, -np.inf], 0, inplace=True)

print("\n‚úÖ Risk Scoring and Interaction Features Created.")
print(df_fe[['credit_score', 'low_credit_risk_flag', 'employment_stability_score']].head())
```

##### 3.3 Categorical Encoding and Data Scaling

```python
# --- Identify Categorical and Numerical Columns ---
# Columns to drop before modeling:
# Target variables ('emi_eligibility', 'max_monthly_emi') and raw input features
X_raw = df_fe.drop(columns=['emi_eligibility', 'max_monthly_emi'])

# Explicitly ensure columns that should be numeric are indeed numeric in X_raw
# This is a robust step in case earlier cleaning was incomplete or missed.
# The problematic columns identified previously were 'age', 'monthly_salary', 'bank_balance'.
# We also include 'monthly_rent', 'credit_score', 'emergency_fund' to be safe, although they were initially float/int,
# to handle potential issues if their dtypes changed due to earlier operations or NaNs.
numeric_cols_to_clean = ['age', 'monthly_salary', 'bank_balance', 'monthly_rent', 'credit_score', 'emergency_fund']
for col in numeric_cols_to_clean:
    if col in X_raw.columns:
        # Convert to string first to handle any mixed types, then use regex to clean malformed numbers,
        # then coerce to numeric.
        X_raw[col] = X_raw[col].astype(str).str.replace(r'(\.0){2,}$', '.0', regex=True)
        X_raw[col] = pd.to_numeric(X_raw[col], errors='coerce')


categorical_cols = X_raw.select_dtypes(include='object').columns
numerical_cols = X_raw.select_dtypes(include=['float64', 'int64']).columns

# --- One-Hot Encoding for Categorical Features ---
X_encoded = pd.get_dummies(X_raw, columns=categorical_cols, drop_first=True)
print("\n‚úÖ Categorical Encoding Applied.")
print(f"Total features after encoding: {X_encoded.shape[1]}")


# --- Impute Missing Values for Numerical Features in X_encoded ---
from sklearn.impute import SimpleImputer

# Create a list of columns in X_encoded that have NaNs
cols_with_nan_in_X_encoded = X_encoded.columns[X_encoded.isnull().any()].tolist()

if cols_with_nan_in_X_encoded:
    print(f"\nüí° Imputing missing values in {len(cols_with_nan_in_X_encoded)} columns: {cols_with_nan_in_X_encoded}")
    # Using median imputation as it's more robust to outliers than mean
    imputer = SimpleImputer(strategy='median')
    X_encoded[cols_with_nan_in_X_encoded] = imputer.fit_transform(X_encoded[cols_with_nan_in_X_encoded])
    print("‚úÖ Missing values imputed.")
else:
    print("\n‚úÖ No missing values found in X_encoded after encoding.")


# --- Scaling Numerical Features ---
# Identify numerical columns in the encoded dataframe (excluding the new binary/engineered features if they were integers)
# We will scale all continuous-like numerical features
# The condition X_encoded[col].nunique() > 50 checks for high cardinality, implying continuous-like data.
features_to_scale = [col for col in X_encoded.columns if X_encoded[col].nunique() > 50 and col not in ['low_credit_risk_flag', 'employment_stability_score']]
scaler = StandardScaler()
X_encoded[features_to_scale] = scaler.fit_transform(X_encoded[features_to_scale])
print("\n‚úÖ Numerical Feature Scaling Applied to continuous variables (e.g., monthly_salary, credit_score).")
```

##### 3.4 Data Splitting (Train-Test-Validation)

```python
# --- Define Target Variables ---
# Classification Target (emi_eligibility) needs to be mapped to numbers
target_mapping = {'Not_Eligible': 0, 'High_Risk': 1, 'Eligible': 2}
y_classification = df_fe['emi_eligibility'].map(target_mapping)
y_regression = df_fe['max_monthly_emi']

X = X_encoded

# 1. Split into Training and (Test + Validation) Sets
X_train, X_temp, y_cls_train, y_cls_temp, y_reg_train, y_reg_temp = train_test_split(
    X, y_classification, y_regression, test_size=0.3, random_state=42, stratify=y_classification
)

# 2. Split (Test + Validation) into Test and Validation Sets
X_val, X_test, y_cls_val, y_cls_test, y_reg_val, y_reg_test = train_test_split(
    X_temp, y_cls_temp, y_reg_temp, test_size=0.5, random_state=42, stratify=y_cls_temp
)

print("\nüìä Data Split Complete:")
print(f"Total Samples: {len(X)}")
print(f"Training Set: {len(X_train)} ({len(X_train)/len(X):.1%})")
print(f"Validation Set: {len(X_val)} ({len(X_val)/len(X):.1%})")
print(f"Test Set: {len(X_test)} ({len(X_test)/len(X):.1%})")

# The data is now ready for Step 4: Machine Learning Model Development.
```

### Step 4: Machine Learning Model Development

##### 4.1 Set Up and Utility Functions

```python
# --- Define Evaluation Functions ---

def evaluate_classification_model(model_name, y_true, y_pred, y_prob):
    """Calculates and returns key classification metrics."""
    metrics = {
        'Model': model_name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision_Macro': precision_score(y_true, y_pred, average='macro', zero_division=0),
        'Recall_Macro': recall_score(y_true, y_pred, average='macro', zero_division=0),
        'F1_Score_Macro': f1_score(y_true, y_pred, average='macro', zero_division=0),
    }
    # ROC AUC for multi-class (One-vs-Rest)
    if y_prob.shape[1] > 2:
         metrics['ROC_AUC_OVR'] = roc_auc_score(y_true, y_prob, multi_class='ovr')
    else:
         metrics['ROC_AUC'] = roc_auc_score(y_true, y_prob[:, 1])

    return metrics

def evaluate_regression_model(model_name, y_true, y_pred):
    """Calculates and returns key regression metrics."""
    metrics = {
        'Model': model_name,
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
        'MAE': mean_absolute_error(y_true, y_pred),
        'R-squared': r2_score(y_true, y_pred)
    }
    return metrics

# --- Placeholder for MLflow Setup (Detailed in Step 5) ---
# import mlflow
# mlflow.set_experiment("EMIPredict_Risk_Assessment")

print("‚úÖ Libraries and evaluation functions loaded.")
```

##### 4.2 Classification Models (EMI Eligibility Prediction)

```python
# --- CLASSIFICATION MODEL TRAINING ---

classification_results = []
X_train_cls = X_train.copy()
X_val_cls = X_val.copy()
y_train_cls = y_cls_train
y_val_cls = y_cls_val

# 1. Logistic Regression (Baseline Model)
print("\nTraining 1/3: Logistic Regression Classifier...")
log_reg = LogisticRegression(max_iter=1000, multi_class='ovr', solver='lbfgs', random_state=42)
log_reg.fit(X_train_cls, y_train_cls)
y_pred_log_reg = log_reg.predict(X_val_cls)
y_prob_log_reg = log_reg.predict_proba(X_val_cls)
classification_results.append(evaluate_classification_model("Logistic Regression", y_val_cls, y_pred_log_reg, y_prob_log_reg))
print("  - Done.")

# 2. Random Forest Classifier (Feature Importance)
print("Training 2/3: Random Forest Classifier...")
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_clf.fit(X_train_cls, y_train_cls)
y_pred_rf = rf_clf.predict(X_val_cls)
y_prob_rf = rf_clf.predict_proba(X_val_cls)
classification_results.append(evaluate_classification_model("Random Forest", y_val_cls, y_pred_rf, y_prob_rf))
print("  - Done.")

# 3. XGBoost Classifier (High-Performance Gradient Boosting)
print("Training 3/3: XGBoost Classifier...")
# We use a balanced scale_pos_weight for potential imbalance in the 3 classes
xgb_clf = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42,
    n_jobs=-1
)
xgb_clf.fit(X_train_cls, y_train_cls)
y_pred_xgb = xgb_clf.predict(X_val_cls)
y_prob_xgb = xgb_clf.predict_proba(X_val_cls)
classification_results.append(evaluate_classification_model("XGBoost Classifier", y_val_cls, y_pred_xgb, y_prob_xgb))
print("  - Done.")

# --- Summary of Classification Results ---
cls_summary_df = pd.DataFrame(classification_results).set_index('Model')
print("\n\nüìä Classification Model Performance on Validation Set:")
print(cls_summary_df.sort_values(by='Accuracy', ascending=False))
```

##### 4.3 Regression Models (Maximum EMI Amount Prediction)

```python
# --- REGRESSION MODEL TRAINING ---

regression_results = []
X_train_reg = X_train.copy()
X_val_reg = X_val.copy()
y_train_reg = y_reg_train
y_val_reg = y_reg_val

# 1. Linear Regression (Baseline Model)
print("\nTraining 1/3: Linear Regression...")
lin_reg = LinearRegression(n_jobs=-1)
lin_reg.fit(X_train_reg, y_train_reg)
y_pred_lin_reg = lin_reg.predict(X_val_reg)
regression_results.append(evaluate_regression_model("Linear Regression", y_val_reg, y_pred_lin_reg))
print("  - Done.")

# 2. Random Forest Regressor (Ensemble-based Predictions)
print("Training 2/3: Random Forest Regressor...")
rf_reg = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_reg.fit(X_train_reg, y_train_reg)
y_pred_rf_reg = rf_reg.predict(X_val_reg)
regression_results.append(evaluate_regression_model("Random Forest Regressor", y_val_reg, y_pred_rf_reg))
print("  - Done.")

# 3. XGBoost Regressor (Advanced Gradient Boosting)
print("Training 3/3: XGBoost Regressor...")
xgb_reg = XGBRegressor(n_estimators=100, learning_rate=0.1, objective='reg:squarederror', random_state=42, n_jobs=-1)
xgb_reg.fit(X_train_reg, y_train_reg)
y_pred_xgb_reg = xgb_reg.predict(X_val_reg)
regression_results.append(evaluate_regression_model("XGBoost Regressor", y_val_reg, y_pred_xgb_reg))
print("  - Done.")

# --- Summary of Regression Results ---
reg_summary_df = pd.DataFrame(regression_results).set_index('Model')
print("\n\nüìä Regression Model Performance on Validation Set:")
# Project target is RMSE below 2000 INR
print(reg_summary_df.sort_values(by='RMSE', ascending=True))
```

##### 5.2 Log All Models with MLflow Tracking

```python
# --- MLflow Logging Function ---

def log_model_to_mlflow(model_dict, model_type):
    """Logs a single model's details, metrics, and artifact to MLflow."""
    model_name = list(model_dict.keys())[0]
    data = model_dict[model_name]

    with mlflow.start_run(run_name=model_name) as run:
        # Log Tags (for easy filtering)
        mlflow.set_tag("model_type", model_type)
        mlflow.set_tag("status", "Validated")

        # Log Parameters
        mlflow.log_params(data["params"])

        # Log Metrics
        mlflow.log_metrics(data["metrics"])

        # Log Model Artifact
        mlflow.sklearn.log_model(
            sk_model=data["model"],
            artifact_path=model_type,
            registered_model_name=f"EMI_{model_type}_{model_name.replace(' ', '_')}"
        )
        print(f"‚úÖ Logged {model_name} (Run ID: {run.info.run_id})")

# Log Classification Models
print("Logging Classification Models:")
for name, data in classification_models.items():
    log_model_to_mlflow({name: data}, "Classification")

# Log Regression Models
print("\nLogging Regression Models:")
for name, data in regression_models.items():
    log_model_to_mlflow({name: data}, "Regression")

print("\nAll models have been logged, tracked, and saved as artifacts.")
```
